## 1-2. 다중 그래프(multiple graphs)
plot multiple graphs within one canvas

```python
#그래프 그려지는 과정
data1 = np.arange(1, 51)
data2 = np.arange(51, 101)

#캔버스에 그래프 올라가는 과정
plt.plot(data1)
plt.plot(data2)

#show 전에 여러개 그래프를 동시에 올릴 수 있다. 그려지는 순간과 
plt.show()
```

2개의  figure 나눠서 다중그래프 그리기
plt.figure() :
 canvas 설정 함수, 호출시 캔버스가 하나씩 만들어짐

```python
plt.plot(data1) # data1 시각화
plt.figure() # 새로운 캔버스 생성
plt.plot(data2) # 새롭게 만들어진 캔버스에 data 2시각화
plt.show() #모두 보야주기
```

## 1-3. 여러개의 plot 그리는 방법 (subplot)
    * subplot(row, column, index) 여기서 index는 1부터 시작



```python
plt.subplot(2, 1, 1) #2개의 행을 만들고 하나의 열을 만들어 첫번째 plot에 표시
plt.plot(data1)

plt.subplot(2,1, 2)
plt.plot(data2)

plt.show()
```

# 위와 똑같은 그래프를 그리지만 다른 표현
    * 콤마(,)를 생략하고 그릴수 있음

```python
plt.subplot(211) # row 2 col 1 index 1
plt.plot(data1) 

plt.subplot(212) # row 2 col 1 index 2
plt.plot(data2)

plt.show()
```

#2-4. legend 설정



범례 따로 넣어주기

```python
plt.plot(data1 * 2)
plt.plot(data1 ** 2)
plt.plot(np.log(data1))

#title 설정
plt.title("x, y tick rotation")

#x, y 축 설정
plt.xlabel("x axies")
plt.ylabel("x axies")

#x, y tick 설정
plt.xticks(rotation=90)
plt.yticks(rotation=30)

#elgend (범례) 설정
plt.legend(["data1 * 2", "data1 **2", "log(data1"])

plt.show()
```

plot과 함께 label 넣어서 범례 설정

```python
plt.plot(data1 * 2, label="a")
plt.plot(data1 ** 2, label="b")
plt.plot(np.log(data1), label="c")

#title 설정
plt.title("x, y tick rotation")

#x, y 축 설정
plt.xlabel("x axies")
plt.ylabel("x axies")

#x, y tick 설정
plt.xticks(rotation=90)
plt.yticks(rotation=30)

#elgend (범례) 설정
plt.legend()

plt.show()
```

#2-5. x와 y 의 한계점
* tick의 한계점 설정


```python
plt.plot(data1 * 2, label="a")
plt.plot(data1 ** 2, label="b")
plt.plot(np.log(data1), label="c")

#title setting
plt.title("x, y tick rotation")

#x, y 축 setting
plt.xlabel("x axies")
plt.ylabel("x axies")

#x, y tick setting
plt.xticks(rotation=90)
plt.yticks(rotation=30)

#elgend (범례) setting
plt.legend()

plt.xlim(0, 10)
plt.ylim(20, 100)


# x, y limitation setting 
plt.show()
```

#2-6. Style detail marker line color setting
 * marker의 종류 외울 필요 없음 필요시 찾아쓰기

```python
data = np.arange(10)

plt.plot(data, data * 2, marker = 'o', markersize =25)
plt.plot(data, data * 2 -10 , marker = 'h', markersize =25)
plt.plot(data, data * 2 - 20, marker = 'd', markersize =15)
plt.plot(data, data * 2 - 30, marker = '*', markersize =35)
plt.plot(data, data * 2 -40 , marker = '^', markersize =25)

plt.show()
```

# line type
'-' solid line style
'--' dashed line style
'-.' dash-dot line style
':' dotted line style

```python
data = np.arange(10)

plt.plot(data, data * 2,      marker = 'o', linestyle = ' ') #선이 안보이고 점만
plt.plot(data, data * 2 -10 , marker = 'h', linestyle = '-') #실선
plt.plot(data, data * 2 - 20, marker = 'd', linestyle = '--') #짧은 실선
plt.plot(data, data * 2 - 30, marker = '*', linestyle = '-.') #선-점
plt.plot(data, data * 2 - 40, marker = '^',linestyle =':')

plt.show()
```

```python
data = np.arange(10)

plt.plot(data, data * 2,      marker = 'o', linestyle = ' ', color ='k') #선이 안보이고 점만
plt.plot(data, data * 2 -10 , marker = 'h', linestyle = '-', color ='c') #실선
plt.plot(data, data * 2 - 20, marker = 'd', linestyle = '--', color ='y') #짧은 실선
plt.plot(data, data * 2 - 30, marker = '*', linestyle = '-.', color ='b') #선-점
plt.plot(data, data * 2 - 40, marker = '^', linestyle =':',  color ='m') 

plt.show()
```

```python
data = np.arange(10)

plt.plot(data, data * 2,      marker = 'o', linestyle = ' ', color ='k', alpha=0.1) #선이 안보이고 점만
plt.plot(data, data * 2 -10 , marker = 'h', linestyle = '-', color ='c', alpha=0.3) #실선
plt.plot(data, data * 2 - 20, marker = 'd', linestyle = '--', color ='y', alpha=0.5) #짧은 실선
plt.plot(data, data * 2 - 30, marker = '*', linestyle = '-.',  color ='b', alpha=0.7) #선-점
plt.plot(data, data * 2 - 40, marker = '^', linestyle = ':',  color ='m', alpha=0.9) #점선

plt.show()
```

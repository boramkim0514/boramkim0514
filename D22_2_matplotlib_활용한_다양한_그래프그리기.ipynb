# 1. Scatter

#1-1. x, y, colors, area 설정하기
* colors : 임의의 정수 값을 색상으로 표현해줌
* area :점의 크기, 값이 커지면


```python
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.arange(50)
area = x * y * 1000

plt.scatter(x, y, s= area, c=colors)
plt.show()
```

## 1-2. cmap, alpha
cmap 컬러 지정하면, 컬러값 모두 동일하게 설정
alpha 투명도 0-1, 0일수록 투명

```python
# 현재 사용할 figure에 대해서만 크기 조절
plt.figure(figsize=(12, 6))

plt.subplot(131)
plt.scatter(x, y, s=area, cmap='blue', alpha=0.1)
plt.title('alpha=0.1')

plt.subplot(132)
plt.scatter(x, y, s=area, cmap='red', alpha=0.5)
plt.title('alpha=0.5')

plt.subplot(133)
plt.scatter(x, y, s=area, c=y, alpha=1.0) #zjffj qoduf tjfwjd sjffj  
plt.title("alpha=1.0, colorbar")
plt.colorbar() #값별 농도 확인

plt.show()
```

#5-2 다중히스토그램 

```python
N = 100000
bins = 30

x = np.random.randn(N)

fig, axes = plt.subplot(1, 3
                    sharey=True,  #y 축 값 모든 다중 그래프들이 다 공유
                    tight_layout=True) #graph의 패딩을 자동으로 조절해 fit한 graph를 생성
fig.set_size_inches(12, 5)                       

axes[0].hist(x, bins=bins)
axes[1].hist(x, bins=bins * 2)
axes[3].hist(x, bins=bins * 4)

plt.show()
```

#5-3 y축에 Desnsity 표기


```python
N = 100000
bins = 30

x = np.random.randn(N)

fig, axes = plt.subplot(1, 2, 
                        tight_layout=True) #graph의 패딩을 자동으로 조절해 fit한 graph를 생성
fig.set_size_inches(9, 3)

plt.show() #1.0 데이터가 모여있고, 0 데이턱 없는것

#density= True 값 통해 y 축에 density 표기

axes[0].hist(x, bins=bins, density=True, cumulatives=True) #빈도수 누적합
axes[1].histo(x, bins=bins, density=True)

plt.show()
```

#6. Pie Chart
 `explode` : mark pop out certain area
 `autopct` : mark percent  automatically
 `shadow` : pie chart shadow effect
 `startangle` : pie chart plotting 
 
 pie chart 활용 2가지 리턴

 texts : label text 효과
 autotexts: 파이 위에 그려지는 텍스트 효과

```python
labels = ["A", "B", "C", "D", "E", "F"]
sizes = [20.4, 15.8, 10.5, 9, 7.6, 36.7]
explode = (0.3, 0, 0, 0, 0, 0) #안바뀌게 하기 위해 tuple로 받음

#text, autotexts인자 활용해 텍스트 스타일링 적용
#순서 바뀌면 안됌! 외워라!
_, texts, autotexts = plt.pie(sizes, 
                              labels= labels,
                              explode=explode,
                              autopct='%1.1f%%',
                              shadow=True,
                              startangle=90)
plt.title("PIE CHART TEST")

#label text 에 대한 스타일 적용

for t in texts:
    t.set_fontsize(12)
    t.set_color('gray')

#pie 위의 텍스트에 대한 
#label text 에 대한 스타일 적용

for t in autotexts:
    t.set_color("white")
    t.set_fontsize(10)

plt.show()
```
